```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
data = pd.read_csv('final_dataset.csv')
```

```python
#let's check data for regions with id 2
data[data["region_id"] == 2]
```

```python
#now let's start to encode by region_id columns
to_encode_by = ['region_id']

encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(data[to_encode_by])

# new df with the encoded columns
# use get_feature_names_out to get the column names for the encoded data
encoded_df_oh = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(to_encode_by))

# concat the encoded dataframe with the original dataframe
encoded = pd.concat([data, encoded_df_oh], axis=1)

# drop region_id column
encoded = encoded.drop(to_encode_by, axis=1)

encoded
```

```python
#check if we have same records for region 2 saved in updated dataset
encoded[encoded["region_id_2"]==1]
```

```python
encoded.to_csv('final_dataset_encoded.csv', index=False)
```
