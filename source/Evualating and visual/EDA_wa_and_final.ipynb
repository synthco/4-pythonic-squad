```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
final_weather_alarms = pd.read_csv('final_weather_alarms.csv')
final_weather_alarms
```

```python
visib_alarm = final_weather_alarms[["hour_visibility", "is_alarm"]]
visib_alarm
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.violinplot(x='is_alarm', y='hour_humidity', data=final_weather_alarms, palette=['deepskyblue', 'green'])
plt.title('Distribution of Humidity by Alarm Status')
plt.xlabel('Alarm Status')
plt.ylabel('Humidity')
plt.xticks([0, 1], ['No Alarm', 'Alarm'])
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='is_alarm', y='hour_cloudcover', data=final_weather_alarms, palette=['pink', 'yellow'])
plt.title('Distribution of Cloud Cover by Alarm Status')
plt.xlabel('Alarm Status')
plt.ylabel('Cloud Cover')
plt.xticks([0, 1], ['No Alarm', 'Alarm'])
plt.show()
```
